```python
import pandas as pd
```

```python
# read in template

input_suppliers = ['SARA', 'AIKHA', 'HERMES']
template = pd.read_excel('supplier_comparison_template_use.xlsx', sheet_name="Supplier Quotation", header=[1,2])


input_suppliers_lower = set(s.lower() for s in input_suppliers)
print(input_suppliers)
template.head()


```

```python
new_columns = []
for col in template.columns:
    if isinstance(col, tuple) and col[0].lower() in input_suppliers_lower:
        new_columns.append('_'.join(col).strip())  # e.g., HERMES_UP
    else:
        new_columns.append(col[0] if isinstance(col, tuple) else col)  # e.g., ITEM CODE, QTY
template.columns = new_columns

print(template.columns)
```

```python
# calculate the total price for each supplier
for supplier in input_suppliers:
    supplier_col = f"{supplier}_UP"
    col_index = template.columns.get_loc(supplier_col)
    if supplier_col in template.columns:
        total_values = template[supplier_col] * template['QTY']
        template.insert(loc=col_index + 1, column=f"{supplier}_TOTAL", value=total_values)
    else:
        print(f"Warning: Column {supplier_col} not found in the template.")
```

```python
blank_row = pd.DataFrame([{col: "" for col in template.columns}])

# Step 2: Create summary row
summary_row = {'ITEM CODE': 'TOTAL_QUOTE'}  # you can also label 'DESCRIPTION' or a new column
for col in template.columns:
    if col.endswith("_TOTAL"):
        summary_row[col] = template[col].sum()

summary_row_df = pd.DataFrame([summary_row])

# Step 3: Concatenate everything
template = pd.concat([template, blank_row, summary_row_df], ignore_index=True)
```
